```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
G = nx.DiGraph()

# Define valid decision-making structure
edges = [
    ("Start", "Government: Repatriate (R)"),
    ("Start", "Government: Do Nothing (N)"),

    # If Government Repatriates
    ("Government: Repatriate (R)", "Public: Protest (P)"),
    ("Government: Repatriate (R)", "Public: Do Nothing (n)"),

    # If Public Protests
    ("Public: Protest (P)", "Extremists: Fail (D)"),

    # If Public Does Nothing after Repatriation
    ("Public: Do Nothing (n)", "Extremists: Fail (D)"),

    # If Government Does Nothing
    ("Government: Do Nothing (N)", "Extremists: Recruit (T)"),
    ("Government: Do Nothing (N)", "Extremists: Fail (D)"),
]

# Add edges to the graph
G.add_edges_from(edges)

# Define a structured hierarchical layout
pos = {
    "Start": (3, 6),
    "Government: Repatriate (R)": (1, 5),
    "Government: Do Nothing (N)": (5, 5),
    "Public: Protest (P)": (0, 4),
    "Public: Do Nothing (n)": (2, 4),
    "Extremists: Recruit (T)": (4, 4),
    "Extremists: Fail (D)": (6, 4),
}

# Terminal nodes with payoff values
payoffs = {
    "Extremists: Recruit (T)",
    "Extremists: Fail (D)"
}

# Draw the graph
plt.figure(figsize=(10, 7))
nx.draw(G, pos, with_labels=True, node_color="lightblue", edge_color="black", node_size=3000, font_size=9, font_weight="bold", arrowsize=15)

# Add payoff labels near terminal nodes
# Add payoff labels near terminal nodes
for node in payoffs:
    if node in pos:
        x, y = pos[node]
        label = payoffs[node] if isinstance(payoffs[node], str) else "\n".join([f"{k}: {v}" for k, v in payoffs[node].items()])
        plt.text(x, y - 0.3, label, fontsize=10, ha="center", color="red", fontweight="bold")

plt.title("Final Decision Tree for UK Repatriation Policy", fontsize=12, fontweight="bold")
plt.show()


```

```python
from graphviz import Digraph

# First tree: Public Response and Government Impact
def create_public_response_tree():
    dot = Digraph(comment='Public Response Decision Tree')
    dot.attr(rankdir='TB')

    # Add nodes
    dot.node('A', 'Government\nDecision')
    dot.node('B', 'Public\nResponse\n(If Repatriate)')
    dot.node('C', 'Public\nResponse\n(If Do Nothing)')

    # Government outcomes when Repatriate + Public Protests
    dot.node('D', 'Gov Payoff: +5\nPublic Payoff: -5')
    # Government outcomes when Repatriate + Public No Action
    dot.node('E', 'Gov Payoff: +10\nPublic Payoff: +10')
    # Government outcomes when No Action + Public Protests
    dot.node('F', 'Gov Payoff: -10\nPublic Payoff: 0')
    # Government outcomes when No Action + Public No Action
    dot.node('G', 'Gov Payoff: +5\nPublic Payoff: +10')

    # Add edges
    dot.edge('A', 'B', 'Repatriate (R)')
    dot.edge('A', 'C', 'Do Nothing (N)')
    dot.edge('B', 'D', 'Protest (P)')
    dot.edge('B', 'E', 'No Action (n)')
    dot.edge('C', 'F', 'Protest (P)')
    dot.edge('C', 'G', 'No Action (n)')

    return dot

# Second tree: Terrorist Recruitment and Government Impact
def create_terrorist_tree():
    dot = Digraph(comment='Terrorist Recruitment Decision Tree')
    dot.attr(rankdir='TB')

    # Add nodes
    dot.node('A', 'Government\nDecision')
    dot.node('B', 'Extremist\nOutcome\n(If Repatriate)')
    dot.node('C', 'Extremist\nOutcome\n(If Do Nothing)')

    # Government outcomes when Repatriate + Successful Recruitment
    dot.node('D', 'Gov Payoff: +5\nExtremist Success in Recruitment: -5')
    # Government outcomes when Repatriate + Unsuccessful Recruitment
    dot.node('E', 'Gov Payoff: +10\nExtremist Success in Recruitment: -10')
    # Government outcomes when No Action + Successful Recruitment
    dot.node('F', 'Gov Payoff: -10\nExtremist Success in Recruitment: +10')
    # Government outcomes when No Action + Unsuccessful Recruitment
    dot.node('G', 'Gov Payoff: +5\nExtremist Success: 0')

    # Add edges
    dot.edge('A', 'B', 'Repatriate (R)')
    dot.edge('A', 'C', 'Do Nothing (N)')
    dot.edge('B', 'D', 'Success (T)')
    dot.edge('B', 'E', 'Failure (D)')
    dot.edge('C', 'F', 'Success (T)')
    dot.edge('C', 'G', 'Failure (D)')

    return dot

# Create and save both trees
public_tree = create_public_response_tree()
terrorist_tree = create_terrorist_tree()

# Save to files
public_tree.render('public_response_tree', format='png', cleanup=True)
terrorist_tree.render('terrorist_recruitment_tree', format='png', cleanup=True)
```
